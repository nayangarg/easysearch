```python
import os
import pandas as pd
import csv
```

```python
wrtr = csv.writer(open('index1.csv','w'))
rdr = csv.reader(open('index.csv','r'))
```

```python
for i in list(rdr)[1:]:
    docs = []
    print(i[0])
    for j in i[1:]:
        
        if j == '':
            break
        rnklst = pd.read_csv(open('/home/nayan/coding/Major/sCake Phrase/venv/SCScore_W/'+j[:-4] + '_ranked_list.csv', 'r'))
        tup = [j]
        for k in range(50):
            if rnklst['Words'][k] == i[0]:
                tup.append(k)
                break
        docs.append(tup)
    lst = [i[0]] + docs
    wrtr.writerow(lst)
```
