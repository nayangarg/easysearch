```python
wrtr = csv.writer(open('index1.csv','w'))
rdr = csv.reader(open('index.csv','r'))
rdr = list(rdr)
```

```python
for i in rdr[6085:]:
    docs = []
    print(i[0])
    for j in i[1:]:
        print(j)
        
        if j == '':
            break
        rnklst = pd.read_csv(open('/home/nayan/coding/Major/sCake Phrase/venv/SCScore_W/'+j[:-4] + '_ranked_list.csv', 'r'))
        print(rnklst.head())
        tup = [j]
        for k in range(50):
            if rnklst['Words'][k] == i[0]:
                tup.append(k)
                print(k)
                break
        docs.append(tup)
        print(docs)
    lst = [i[0]] + docs
    print(lst)
    wrtr.writerow(lst)
```
