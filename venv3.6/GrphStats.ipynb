```python
import pandas as pd
import networkx as nx
```

```python
import csv
import os
```

```python
data_path = '/home/nayan/coding/Major/sCake Phrase/venv/edgelists'
nodes = 0
files = 0
edges = 0
cc = 0
apl = 0
den = 0
```

```python
for every_file in (os.listdir(data_path)):
#     if every_file != '316184.csv':
#         continue
    #print(every_file)
    rdr = csv.reader(open(data_path+'/'+every_file,'r'),delimiter = '\t')
    G = nx.Graph()
    for i in list(rdr):
        G.add_edge(i[0],i[1])
    nodes = nodes + G.number_of_nodes()
    edges = edges + G.number_of_edges()
    cc = cc + nx.transitivity(G)
    for C in (G.subgraph(c).copy() for c in nx.connected_components(G)): 
        apl = apl + nx.average_shortest_path_length(C)    
    den = den + nx.density(G)
```

```python
print(nodes/2304)
print(edges/2304)
print(cc/2304)
print(apl/2304)
print(den/2304)
```
