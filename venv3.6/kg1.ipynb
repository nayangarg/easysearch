```python
import csv
import networkx as nx
import os
import pandas as pd
import math
```

```python
G = nx.Graph()
```

```python
edg = dict()
```

```python
class node:
    def __init__(self):
        self.pos = []
        self.sen = []
        self.l = 0
```

```python
data_path = '/home/nayan/coding/Major/sCake Phrase/venv/wordList'
```

```python
for every_file in (os.listdir(data_path)):
    if every_file == 'combined.csv':
        continue
    rdr = csv.reader(open(data_path + '/' + every_file, 'r'))
    df = pd.read_csv(open('/home/nayan/coding/Major/sCake Phrase/venv/SCScore_W/'+every_file[:-4] + '_ranked_list.csv', 'r'))
    key = list(df['Words'][:50])
    nodes = dict() 
    
    for i in range(50):
        if not G.has_node(key[i]):
            G.add_node(key[i])
        nodes[key[i]] = node()
        if key[i] not in edg:
            edg[key[i]] = dict()
        
    p = 0
    s = 0

    for line in rdr:
        for i in line:
            if i in key:
                nodes[i].pos.append(p)
                p += 1
                nodes[i].sen.append(s)
        s += 1
    for i in nodes:
        nodes[i].l = len(nodes[i].pos)
    
    for i in key:
        for j in key:
            
            if i == j:
                continue
            
            wt = 0
            
            for a in range(nodes[i].l):
                for b in range(nodes[j].l):
                    wt += ((nodes[i].pos[a] - nodes[j].pos[b])**2) * ((nodes[i].sen[a] - nodes[j].sen[b] + 1)**2)
            #print(i,j,wt)
            
            if G.has_edge(i, j):
                G[i][j]['weight'] += wt/2
            else:
                G.add_edge(i, j, weight = wt/2)
                
            if j in edg[i]:
                edg[i][j] += (nodes[i].l*nodes[j].l)
            else:
                edg[i][j] = (nodes[i].l*nodes[j].l)
                
```

```python
for (u, v, d) in G.edges().data():
    if edg[u][v] == 0:
        edg[u][v] = 1
    G[u][v]['weight'] /= edg[u][v]
    G[u][v]['weight'] = math.sqrt(G[u][v]['weight'])
```

```python
print(sorted(G.edges(data=True), key=lambda t: t[2].get('weight', 1)))
```
